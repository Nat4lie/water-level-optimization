```
# Now plot using plotly express
import plotly.express as px
for lake in df.columns[2:]:
    fig = px.line(df, x="Month", y=lake, color="Year", title='Water levels of ' + lake + ' in three years', labels={"Month": "Month", "variable": "Water level", "value": lake})
    fig.update_layout(
    title={
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'})
    fig.show()
```
